# Sistema de Ranking de Alunos

Este exercício aborda:
- Listas e dicionários
- Ordenação (Bubble, Insertion, Selection)
- Busca Linear e Binária
- Laços
- Operador ternário


```python
# ==============
# Base de Dados
# ==============

alunos = []

def cadastrar_aluno(id, nome : str, nota : float):
    aluno = {
        "id": id,
        "nome": nome,
        "nota": nota
    }
    alunos.append(aluno)

cadastrar_aluno(1,"Neto",5)
cadastrar_aluno(1,"Neto",6)
cadastrar_aluno(1,"Neto",7)
cadastrar_aluno(1,"Neto",8)
cadastrar_aluno(1,"Neto",9)
cadastrar_aluno(1,"Neto",10)
print(alunos)
```

```python
# =============
# Busca Linear
# =============

def busca_linear(id : int):
    for aluno in alunos:
        if aluno["id"] == id:
            return aluno
    return None

if busca_linear(2) is None:
  print("Não encontrou. ")
```

```python
# =======================================
# Busca Binária (requer ordenação por ID)
# =======================================

def busca_binaria(id):
    esquerda = 0
    direita = len(alunos) - 1

    while esquerda <= direita:
        meio = (esquerda + direita) // 2

        if alunos[meio]["id"] == id:
            return alunos[meio]
        elif alunos[meio]["id"] < id:
            esquerda = meio + 1
        else:
            direita = meio - 1

    return None

busca_binaria(1)
```

```python
# ==================================
# Bubble Sort (por nota decrescente)
# ==================================

def bubble_sort():
    n = len(alunos)

    for i in range(n):
        for j in range(0, n - i - 1):
            if alunos[j]["nota"] < alunos[j + 1]["nota"]:
                alunos[j], alunos[j + 1] = alunos[j + 1], alunos[j]
```

```python
# ===============
# Insertion Sort
# ===============

def insertion_sort():
    for i in range(1, len(alunos)):
        print(f"Valor do i: {i}")
        chave = alunos[i]
        print(f"Valor de chave é: {chave['nota']}")
        j = i - 1
        print(f"Valor de j: {j}")

        print(f"Comparando {alunos[j]["nota"]} menor que {chave["nota"]}")
        while j >= 0 and alunos[j]["nota"] < chave["nota"]:
            alunos[j + 1] = alunos[j]
            print(f"alunos[{j+1}] recebeu {alunos[j]['nota']}")
            print(alunos)
            j -= 1
            print(f"Valor de j é : {j}")

        alunos[j + 1] = chave
        print(alunos)
        print(f"alunos[{[j + 1]}] recebeu a chave que era {chave['nota']}")

insertion_sort()
```

```python
# ================
# Selection Sort
# ================

def selection_sort():
    n = len(alunos)

    for i in range(n):
        max_idx = i

        for j in range(i + 1, n):
            if alunos[j]["nota"] > alunos[max_idx]["nota"]:
                max_idx = j

        alunos[i], alunos[max_idx] = alunos[max_idx], alunos[i]
```

```python
# ===============
# Classificação
# ===============

def classificar(aluno):
    return "Aprovado" if aluno["nota"] >= 6 else "Reprovado"
```

```python
# ================
# Exibir Ranking
# ================

def mostrar_ranking():
    print("\n=== RANKING ===")
    for i, aluno in enumerate(alunos):
        status = classificar(aluno)
        print(f"{i+1}º - {aluno['nome']} | Nota: {aluno['nota']} | {status}")
```

```python
def menu():
    print("\n===== MENU =====")
    print("1 - Cadastrar aluno")
    print("2 - Listar alunos")
    print("3 - Ordenar (Bubble Sort)")
    print("4 - Ordenar (Insertion Sort)")
    print("5 - Ordenar (Selection Sort)")
    print("6 - Busca Linear")
    print("7 - Busca Binária")
    print("0 - Sair")
```

```python
def sistema():
    while True:
        menu()
        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            id = int(input("ID: "))
            nome = input("Nome: ")
            nota = float(input("Nota: "))
            cadastrar_aluno(id, nome, nota)
            print("Aluno cadastrado!")

        elif opcao == "2":
            mostrar_ranking()

        elif opcao == "3":
            bubble_sort()
            print("Ordenado com Bubble Sort.")

        elif opcao == "4":
            insertion_sort()
            print("Ordenado com Insertion Sort.")

        elif opcao == "5":
            selection_sort()
            print("Ordenado com Selection Sort.")

        elif opcao == "6":
            id = int(input("Digite o ID: "))
            resultado = busca_linear(id)
            print(resultado if resultado else "Aluno não encontrado.")

        elif opcao == "7":
            print("Ordenando por ID para busca binária...")
            alunos.sort(key=lambda x: x["id"])
            id = int(input("Digite o ID: "))
            resultado = busca_binaria(id)
            print(resultado if resultado else "Aluno não encontrado.")

        elif opcao == "0":
            print("Encerrando sistema...")
            break

        else:
            print("Opção inválida!")
```

```python
sistema()
```
