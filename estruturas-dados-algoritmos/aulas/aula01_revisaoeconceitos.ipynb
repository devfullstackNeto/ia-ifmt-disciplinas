# üîµ PARTE 1 ‚Äì LISTAS (VETORES)

```python
numeros = [10, 20, 30, 40]
print(numeros)
```

## üìå Acesso direto (O(1))

```python
print(numeros[2])
```

## üîÅ Percorrendo Lista (O(n))

```python
for valor in numeros:
    print(valor)
```

## ‚ûï Inser√ß√£o e Remo√ß√£o
- append() ‚Üí O(1) amortizado
- insert() ‚Üí O(n)
- pop() ‚Üí O(1) no final

```python
numeros.append(50)
numeros.insert(1, 15)
numeros.pop()
print(numeros)
```

## üîé Busca Linear (O(n))

```python
def busca_linear(lista, alvo):
    for i in range(len(lista)):
        if lista[i] == alvo:
            return i
    return -1

print(busca_linear(numeros, 30))
```

## üî• List Comprehension

```python
pares = [x for x in range(20) if x % 2 == 0]
print(pares)
```

# üü¢ PARTE 2 ‚Äì DICION√ÅRIOS (HASH TABLES)

```python
aluno = {
    "nome": "Ana",
    "idade": 20,
    "curso": "IA"
}

print(aluno["nome"])
```

## üîÅ Frequ√™ncia de Palavras (Aplica√ß√£o Hash)

```python
frequencia = {}
palavras = ["ia", "dados", "ia", "python", "dados"]

for p in palavras:
    if p in frequencia:
        frequencia[p] += 1
    else:
        frequencia[p] = 1

print(frequencia)
```

# üü£ PARTE 3 ‚Äì TUPLAS

# üü£ FILAS (FIFO)

```python
from collections import deque

fila = deque()
fila.append("A")
fila.append("B")
fila.popleft()

print(fila)
```

# üå≥ √ÅRVORES (estrutura b√°sica)

```python
class No:
    def __init__(self, valor):
        self.valor = valor
        self.esq = None
        self.dir = None

raiz = No(10)
raiz.esq = No(5)
raiz.dir = No(15)

print(raiz.valor)
```

# üî∑ GRAFOS (lista de adjac√™ncia)

```python
grafo = {
    "A": ["B", "C"],
    "B": ["A", "D"],
    "C": ["A"],
    "D": ["B"]
}

print(grafo["A"])
```
