# PARTE 2 – BUBBLE SORT

## Passo 1 – Comparar dois números

```python
def comparar(a,b):
    if a>b:
        return b,a
    return a,b

print(comparar(5,2))
```

## Passo 2 – Comparar vizinhos

```python
def uma_passagem(lista):
    for i in range(len(lista)-1):
        if lista[i]>lista[i+1]:
            lista[i],lista[i+1]=lista[i+1],lista[i]
    return lista

print(uma_passagem([5,2,9,1]))
```

## Passo 3 – Repetir várias vezes

```python
def bubble_sort(lista):
    n=len(lista)
    for _ in range(n):
        for i in range(len(lista)-1):
            if lista[i]>lista[i+1]:
                lista[i],lista[i+1]=lista[i+1],lista[i]
    return lista

print(bubble_sort([5,2,9,1]))
```

## Passo 4 – Bubble completo otimizado

```python
def bubble_sort_otimizado(lista):
    n=len(lista)
    for i in range(n):
        for j in range(0,n-i-1):
            if lista[j]>lista[j+1]:
                lista[j],lista[j+1]=lista[j+1],lista[j]
    return lista

print(bubble_sort_otimizado([5,2,9,1]))
```

# INSERTION SORT

```python
def insertion_sort(lista):
    for i in range(1,len(lista)):
        chave=lista[i]
        j=i-1
        while j>=0 and lista[j]>chave:
            lista[j+1]=lista[j]
            j-=1
        lista[j+1]=chave
    return lista

print(insertion_sort([5,2,9,1]))
```

```python
def insertion_sort_passos(lista):
    print("Lista inicial:", lista)
    print("-" * 40)

    for i in range(1, len(lista)):
        chave = lista[i]
        j = i - 1

        print(f"\n Iteração i={i}")
        print(f"Chave escolhida = {chave}")

        while j >= 0 and lista[j] > chave:
            print(f"Comparando {lista[j]} > {chave} → True")
            print(f"Movendo {lista[j]} para posição {j+1}")

            lista[j + 1] = lista[j]
            print("Estado da lista:", lista)

            j -= 1

        print(f"Inserindo chave {chave} na posição {j+1}")
        lista[j + 1] = chave
        print("Lista após inserção:", lista)
        print("-" * 40)

    print("\nLista ordenada:", lista)
    return lista


insertion_sort_passos([5,2,9,1])
```

# PARTE 5 – MERGE SORT PASSO A PASSO

```python
def merge(esq,dir):
    r=[]
    i=j=0
    while i<len(esq) and j<len(dir):
        if esq[i]<dir[j]:
            r.append(esq[i]); i+=1
        else:
            r.append(dir[j]); j+=1
    r+=esq[i:]; r+=dir[j:]
    return r

def merge_sort(lista):
    if len(lista)<=1:
        return lista
    m=len(lista)//2
    return merge(merge_sort(lista[:m]),merge_sort(lista[m:]))

print(merge_sort([5,2,9,1]))
```

# PARTE 6 – QUICK SORT

```python
def quick_sort(lista):
    if len(lista)<=1:
        return lista
    p=lista[len(lista)//2]
    men=[x for x in lista if x<p]
    ig=[x for x in lista if x==p]
    mai=[x for x in lista if x>p]
    return quick_sort(men)+ig+quick_sort(mai)

print(quick_sort([5,2,9,1]))
```

# PARTE 7 – BUSCAS

```python
def busca_linear(lista,valor):
    for i in range(len(lista)):
        if lista[i]==valor:
            return i
    return -1

print(busca_linear([5,2,9,1],9))
```

```python
def busca_binaria(lista,valor):
    ini,fim=0,len(lista)-1
    while ini<=fim:
        meio=(ini+fim)//2
        if lista[meio]==valor:
            return meio
        elif lista[meio]<valor:
            ini=meio+1
        else:
            fim=meio-1
    return -1

print(busca_binaria(sorted([5,2,9,1]),9))
```

```python
import heapq

def dijkstra(grafo, inicio):
    distancias = {no: float('inf') for no in grafo}
    distancias[inicio] = 0
    fila = [(0, inicio)]

    while fila:
        dist_atual, no_atual = heapq.heappop(fila)

        for vizinho, peso in grafo[no_atual].items():
            distancia = dist_atual + peso

            if distancia < distancias[vizinho]:
                distancias[vizinho] = distancia
                heapq.heappush(fila, (distancia, vizinho))

    return distancias

print(dijkstra(grafo, 'A'))
```

```python
def dijkstra_caminho(grafo, inicio):
    distancias = {no: float('inf') for no in grafo}
    anterior = {no: None for no in grafo}

    distancias[inicio] = 0
    fila = [(0, inicio)]

    while fila:
        dist_atual, no_atual = heapq.heappop(fila)

        for vizinho, peso in grafo[no_atual].items():
            distancia = dist_atual + peso

            if distancia < distancias[vizinho]:
                distancias[vizinho] = distancia
                anterior[vizinho] = no_atual
                heapq.heappush(fila, (distancia, vizinho))

    return distancias, anterior
```

```python
def dijkstra_debug(grafo, inicio):
    import heapq

    distancias = {no: float('inf') for no in grafo}
    distancias[inicio] = 0
    fila = [(0, inicio)]

    passo = 1

    while fila:
        print(f"\nPASSO {passo}")
        print("Fila:", fila)

        dist_atual, no_atual = heapq.heappop(fila)
        print("Visitando:", no_atual, "Distância:", dist_atual)

        for vizinho, peso in grafo[no_atual].items():
            distancia = dist_atual + peso
            print(f" Testando caminho {no_atual}->{vizinho} = {distancia}")

            if distancia < distancias[vizinho]:
                print("  Atualizando!")
                distancias[vizinho] = distancia
                heapq.heappush(fila, (distancia, vizinho))

        print("Distâncias:", distancias)
        passo += 1

    return distancias
```

```python
def caminho(anterior, destino):
    rota = []
    while destino:
        rota.append(destino)
        destino = anterior[destino]
    return rota[::-1]
```

```python
from IPython.display import clear_output
import matplotlib.pyplot as plt
import time

def desenhar(G, pos, atual=None, visitados=set(), caminho_final=None):
    clear_output(wait=True)   # limpa gráfico anterior
    plt.figure(figsize=(5,4))

    cores = []
    for no in G.nodes():
        if caminho_final and no in caminho_final:
            cores.append("green")
        elif no == atual:
            cores.append("red")
        elif no in visitados:
            cores.append("lightblue")
        else:
            cores.append("gray")

    nx.draw(G, pos, with_labels=True, node_color=cores, node_size=800)
    labels = nx.get_edge_attributes(G,'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

    plt.show()
    time.sleep(1)
```

```python
def dijkstra_animado(grafo, inicio):
    import heapq
    dist = {no: float('inf') for no in grafo}
    anterior = {no: None for no in grafo}
    dist[inicio] = 0
    fila = [(0, inicio)]
    visitados = set()

    while fila:
        d, atual = heapq.heappop(fila)
        visitados.add(atual)

        desenhar(G, pos, atual, visitados)

        for vizinho, peso in grafo[atual].items():
            nova = d + peso
            if nova < dist[vizinho]:
                dist[vizinho] = nova
                anterior[vizinho] = atual
                heapq.heappush(fila, (nova, vizinho))

    return dist, anterior
```
