# 1. ALGORITMOS DE ORDENAÇÃO

## Bubble Sort

Compara elementos adjacentes e troca se estiverem fora de ordem.

```python
def bubble_sort(lista):
    n = len(lista)
    for i in range(n):
        for j in range(0,n-i-1):
            if lista[j] > lista[j+1]:
                lista[j],lista[j+1] = lista[j+1],lista[j]
    return lista
```

## Insertion Sort

Insere cada elemento na posição correta.

```python
def insertion_sort(lista):
    for i in range(1,len(lista)):
        chave = lista[i]
        j = i-1
        while j >=0 and chave < lista[j]:
            lista[j+1] = lista[j]
            j -= 1
        lista[j+1] = chave
    return lista
```

## Selection Sort

Seleciona o menor elemento e coloca na posição correta.

```python
def selection_sort(lista):
    n = len(lista)
    for i in range(n):
        menor = i
        for j in range(i+1,n):
            if lista[j] < lista[menor]:
                menor = j
        lista[i],lista[menor] = lista[menor],lista[i]
    return lista
```

# 3. PILHAS

Estrutura LIFO (Last In First Out).

Exemplos reais:

- histórico do navegador
- função desfazer (undo)
- execução de funções na pilha do programa

```python
class Pilha:
    def __init__(self):
        self.itens = []

    def push(self,item):
        self.itens.append(item)

    def pop(self):
        return self.itens.pop()

    def peek(self):
        return self.itens[-1]

    def is_empty(self):
        return len(self.itens) == 0

    def size(self):
        return len(self.itens)

    def clear(self):
        self.itens.clear()

pilha = Pilha()
pilha.push(10)
pilha.push(20)
elemento = pilha.pop()
print(pilha.peek())

print("cadê")
print(elemento)

if (pilha.is_empty()):
  print("Vazia")
```

## Exemplo: histórico do navegador

```python
historico = Pilha()

historico.push("Google")
historico.push("YouTube")
historico.push("GitHub")

print("Página atual:",historico.pop())
```

# 4. FILAS

Estrutura FIFO (First In First Out).

Exemplos reais:

- fila de banco
- fila de hospital
- fila de impressão
- processamento de tarefas

```python
class Fila:
    def __init__(self):
        self.itens = []

    def enqueue(self,item):
        self.itens.append(item)

    def dequeue(self):
        return self.itens.pop(0)

    def front(self):
        return self.itens[0]

    def is_empty(self):
        return len(self.itens) == 0

    def size(self):
        return len(self.itens)

    def clear(self):
        self.itens.clear()
```

## Exemplo: fila de atendimento

```python
fila = Fila()

fila.enqueue("Ana")
fila.enqueue("Carlos")
fila.enqueue("João")

print("Primeiro da fila:",fila.front())

print("Atendendo:",fila.dequeue())
```

**Exercício Prático – Sistema de Gerenciamento com Pilha**

**Contexto:**
Você deverá desenvolver uma pequena aplicação em Python baseada em estruturas de dados utilizando listas, simulando um sistema de gerenciamento de elementos.
O programa deverá funcionar através de um menu interativo, permitindo ao usuário executar diversas operações sobre os dados armazenados.
A aplicação deverá demonstrar conceitos como:<br>
-Pilha (LIFO)
-Inserção em lista
-Remoção de elementos
-Busca
-Ordenação
-Inserção com deslocamento de elementos
-Limpeza da estrutura

**Crie um programa que apresente continuamente o seguinte menu de opções:**
1 - Inserir elemento (push)
2 - Remover elemento (pop)
3 - Mostrar pilha
4 - Buscar elemento
5 - Ordenar elementos
6 - Inserir elemento na posição 2
7 - Inserir no início
8 - Inserir no final
9 - Limpar estrutura
0 - Sair

A aplicação deve continuar executando até que o usuário escolha 0 - Sair.

```python
class Pilha:
    def __init__(self):
        self.itens = []

    def push(self,item):
        self.itens.append(item)

    def pop(self):
        return self.itens.pop()

    def peek(self):
        return self.itens[-1]

    def is_empty(self):
        return len(self.itens) == 0

    def size(self):
        return len(self.itens)

    def clear(self):
        self.itens.clear()


def menu():
    print("\n===== MENU =====")
    print("1 - Inserir elemento (push)")
    print("2 - Remover elemento (pop)")
    print("3 - Mostrar pilha")
    print("4 - Buscar elemento")
    print("5 - Ordenar elementos")
    print("6 - Inserir elemento na posição 2")
    print("7 - Inserir no início")
    print("8 - Inserir no final")
    print("9 - Limpar estrutura")
    print("10 - Mostrar topo da pilha")
    print("11 - Mostrar tamanho da pilha")
    print("0 - Sair")


pilha = Pilha()

while True:

    menu()
    opcao = input("Escolha uma opção: ")

    # PUSH
    if opcao == "1":
        valor = int(input("Digite o valor: "))
        pilha.push(valor)
        print("Elemento inserido!")

    # POP
    elif opcao == "2":
        if pilha.is_empty():
            print("Pilha vazia!")
        else:
            removido = pilha.pop()
            print("Elemento removido:", removido)

    # MOSTRAR
    elif opcao == "3":
        print("Pilha:", pilha.itens)

    # BUSCAR
    elif opcao == "4":
        valor = int(input("Digite o valor a buscar: "))

        if valor in pilha.itens:
            pos = pilha.itens.index(valor)
            print("Elemento encontrado na posição:", pos)
        else:
            print("Elemento não encontrado")

    # ORDENAR
    elif opcao == "5":
        pilha.itens.sort()
        print("Pilha ordenada:", pilha.itens)

    # INSERIR NA POSIÇÃO 2
    elif opcao == "6":
        valor = int(input("Digite o valor: "))

        if pilha.size() >= 2:
            pilha.itens.insert(2, valor)
        else:
            pilha.push(valor)

        print("Elemento inserido na posição 2")
        print(pilha.itens)

    # INSERIR NO INÍCIO
    elif opcao == "7":
        valor = int(input("Digite o valor: "))
        pilha.itens.insert(0, valor)
        print("Inserido no início")
        print(pilha.itens)

    # INSERIR NO FINAL
    elif opcao == "8":
        valor = int(input("Digite o valor: "))
        pilha.push(valor)
        print("Inserido no final")
        print(pilha.itens)

    # LIMPAR
    elif opcao == "9":
        pilha.clear()
        print("Pilha limpa")

    # TOPO
    elif opcao == "10":
        if pilha.is_empty():
            print("Pilha vazia")
        else:
            print("Topo da pilha:", pilha.peek())

    # TAMANHO
    elif opcao == "11":
        print("Tamanho da pilha:", pilha.size())

    # SAIR
    elif opcao == "0":
        print("Encerrando...")
        break

    else:
        print("Opção inválida")
```
