# PARTE 2 – BUBBLE SORT

## Passo 1 – Comparar dois números

```python
def comparar(a,b):
    if a>b:
        return b,a
    return a,b

print(comparar(5,2))
```

## Passo 2 – Comparar vizinhos

```python
def uma_passagem(lista):
    for i in range(len(lista)-1):
        if lista[i]>lista[i+1]:
            lista[i],lista[i+1]=lista[i+1],lista[i]
    return lista

print(uma_passagem([5,2,9,1]))
```

## Passo 3 – Repetir várias vezes

```python
def bubble_sort(lista):
    n=len(lista)
    for _ in range(n):
        for i in range(len(lista)-1):
            if lista[i]>lista[i+1]:
                lista[i],lista[i+1]=lista[i+1],lista[i]
    return lista

print(bubble_sort([5,2,9,1]))
```

## Passo 4 – Bubble completo otimizado

```python
def bubble_sort_otimizado(lista):
    n=len(lista)
    for i in range(n):
        for j in range(0,n-i-1):
            if lista[j]>lista[j+1]:
                lista[j],lista[j+1]=lista[j+1],lista[j]
    return lista

print(bubble_sort_otimizado([5,2,9,1]))
```

# PARTE 3 – SELECTION SORT

```python
def encontrar_menor(lista):
    menor=lista[0]
    for x in lista:
        if x<menor:
            menor=x
    return menor

print(encontrar_menor([5,2,9,1]))
```

```python
def selection_sort(lista):
    for i in range(len(lista)):
        menor=i
        for j in range(i,len(lista)):
            if lista[j]<lista[menor]:
                menor=j
        lista[i],lista[menor]=lista[menor],lista[i]
    return lista

print(selection_sort([5,2,9,1]))
```

# PARTE 4 – INSERTION SORT

```python
def insertion_sort(lista):
    for i in range(1,len(lista)):
        chave=lista[i]
        j=i-1
        while j>=0 and lista[j]>chave:
            lista[j+1]=lista[j]
            j-=1
        lista[j+1]=chave
    return lista

print(insertion_sort([5,2,9,1]))
```

# PARTE 5 – MERGE SORT PASSO A PASSO

```python
def merge(esq,dir):
    r=[]
    i=j=0
    while i<len(esq) and j<len(dir):
        if esq[i]<dir[j]:
            r.append(esq[i]); i+=1
        else:
            r.append(dir[j]); j+=1
    r+=esq[i:]; r+=dir[j:]
    return r

def merge_sort(lista):
    if len(lista)<=1:
        return lista
    m=len(lista)//2
    return merge(merge_sort(lista[:m]),merge_sort(lista[m:]))

print(merge_sort([5,2,9,1]))
```

# PARTE 6 – QUICK SORT

```python
def quick_sort(lista):
    if len(lista)<=1:
        return lista
    p=lista[len(lista)//2]
    men=[x for x in lista if x<p]
    ig=[x for x in lista if x==p]
    mai=[x for x in lista if x>p]
    return quick_sort(men)+ig+quick_sort(mai)

print(quick_sort([5,2,9,1]))
```

# PARTE 7 – BUSCAS

```python
def busca_linear(lista,valor):
    for i in range(len(lista)):
        if lista[i]==valor:
            return i
    return -1

print(busca_linear([5,2,9,1],9))
```

```python
def busca_binaria(lista,valor):
    ini,fim=0,len(lista)-1
    while ini<=fim:
        meio=(ini+fim)//2
        if lista[meio]==valor:
            return meio
        elif lista[meio]<valor:
            ini=meio+1
        else:
            fim=meio-1
    return -1

print(busca_binaria(sorted([5,2,9,1]),9))
```

# 📊 PARTE 8 – GRÁFICOS DE TEMPO

```python
def gerar_lista(n):
    return [random.randint(1,1000) for _ in range(n)]

tamanhos=[100,500,1000]
tb=[]
tq=[]

for n in tamanhos:
    l=gerar_lista(n)
    ini=time.time(); bubble_sort_otimizado(l.copy()); tb.append(time.time()-ini)
    ini=time.time(); quick_sort(l.copy()); tq.append(time.time()-ini)

plt.plot(tamanhos,tb,label='Bubble')
plt.plot(tamanhos,tq,label='Quick')
plt.legend()
plt.show()
```
