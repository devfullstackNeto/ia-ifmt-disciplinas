# LISTAS

```python
numeros=[10,20,30]
print(numeros)
```

Listas são **ordenadas e mutáveis**.

```python
numeros=[10,20,30]
numeros[1]=50
print(numeros)
```

Acessando posições.

```python
numeros=[10,20,30]
print(numeros[0])
```

Python começa no índice 0.

```python
numeros=[5,10,15]
for n in numeros:
 print(n)
```

Adicionar elemento.

```python
numeros=[1,2,3]
numeros.append(4)
print(numeros)
```

Remover elemento.

```python
numeros=[1,2,3,4]
numeros.remove(3)
print(numeros)
```

Tamanho da lista.

```python
numeros=[1,2,3,4]
print(len(numeros))
```

Soma de valores.

```python
numeros=[5,10,15]
print(sum(numeros))
```

Maior valor.

```python
numeros=[3,8,2]
print(max(numeros))
```

Menor valor.

```python
numeros=[3,8,2]
print(min(numeros))
```

Criando lista de notas.

```python
notas=[7,8,6,9]
print(notas)
```

Calculando média.

```python
notas=[7,8,6,9]
media=sum(notas)/len(notas)
print(media)
```

Filtrando números pares.

```python
nums=[1,2,3,4,5,6]
pares=[n for n in nums if n%2==0]
print(pares)
```

Gerando números aleatórios.

```python
import random
nums=[random.randint(1,100) for i in range(10)]
print(nums)
```

Separando pares.

```python
pares=[n for n in nums if n%2==0]
print(pares)
```

# DICIONÁRIOS

```python
aluno={"nome":"Ana","idade":20}
print(aluno)
```

Estrutura chave-valor.

```python
print(aluno["nome"])
```

Adicionando valores.

```python
aluno["nota"]=9
print(aluno)
```

Percorrendo dicionário.

```python
for k,v in aluno.items():
 print(k,v)
```

Lista de dicionários.

```python
alunos=[
{"nome":"Ana","nota":8},
{"nome":"Carlos","nota":7},
{"nome":"Maria","nota":9}
]
print(alunos)
```

Percorrendo registros.

```python
for a in alunos:
 print(a["nome"],a["nota"])
```

Calculando média.

```python
soma=0
for a in alunos:
 soma+=a["nota"]
print(soma/len(alunos))
```

# Aplicação em Ciência de Dados

```python
import pandas as pd
df=pd.DataFrame(alunos)
df
```

DataFrame criado a partir de lista de dicionários.

```python
df.describe()
```

Visualizando média.

```python
df.mean(numeric_only=True)
```

Aluno com maior nota.

```python
df.sort_values("nota",ascending=False).head(1)
```

## Conclusão
