```python
numero = int(input("Digite um número"))
maior = menor = numero

n = 1
while n < 10:
  numero = int(input("Digite um número"))
  if numero > maior:
    maior = numero
  if numero < menor:
    menor = numero
  n += 1

print(f"Maior: {maior:.2f} Menor: {menor:04d}")
```

Questão do Par/Ímpar

```python
par = 0
impar = 0
while True:
  numero = int(input("Digite um número: "))
  if numero == 0:
    break
  if numero % 2 == 0:
    par += 1
  else:
    impar += 1
print(f"Pares: {par} - Ímpares: {impar}")
```

```python
numero = int(input("Digite um número"))
fatorial = numero
for i in range(numero-1, 1, -1):
  fatorial *= i
print(f"{numero}! = {fatorial}")
```

```python
numero = int(input("Digite um número: "))
print(f"{numero}! = {numero}", end="")
fatorial = numero
for i in range(numero-1, 0, -1):
  print(f" x {i}", end="")
  fatorial *= i
print(f" = {fatorial}")
```

```python
a, b = 60, 24
ao, bo = a, b
mdc = 1
d = 2
while a != 1 or b != 1:
  if a % d == 0 and b % d == 0:
    mdc *= d
    print(f"MDC = {mdc}")
    print(f"D = {d}")
  if a % d == 0:
    a //= d
  if b % d == 0:
    b //= d
  if a % d != 0 and b % d != 0:
    d += 1

print(f"MDC de {ao} e {bo} = {mdc}")
```
