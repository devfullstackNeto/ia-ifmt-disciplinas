**Números Kaprekar**



```python
n = int(input("Número: "))

q = n * n
d = len(str(n))

parte1 = q // (10 ** d)
parte2 = q % (10 ** d)

if parte1 + parte2 == n:
    print("É Kaprekar")
else:
    print("Não é Kaprekar")
```

```python
n = int(input("Número: "))
q = n * n

div = 10
kaprekar = False

while div <= q:
    esquerda = q // div
    print(f"Esquerda: {esquerda}")
    direita = q % div
    print(f"Direita: {direita}")

    if direita > 0 and esquerda + direita == n:
        kaprekar = True
        break

    div *= 10

if kaprekar:
    print("É Kaprekar")
else:
    print("Não é Kaprekar")
```

```python
n = int(input("Número: "))
q = n * n

print("n² =", q)

div = 10

while div <= q:
    esquerda = q // div
    direita = q % div

    # inverter a parte direita
    inv = 0
    temp = direita
    while temp > 0:
        inv = inv * 10 + temp % 10
        temp //= 10

    print("\nDivisão:")
    print(esquerda, "|", direita)

    print("Teste 1:", esquerda, "+", direita, "=", esquerda + direita)
    print("Teste 2:", esquerda, "+", inv, "=", esquerda + inv)

    if esquerda + direita == n or esquerda + inv == n:
        print(">>> Encontrou solução!")

    div *= 10
```

```python
n = int(input("Número: "))
q = n * n

print("n² =", q)

digitos = []
temp = q
while temp > 0:
    digitos.append(temp % 10)
    temp //= 10

digitos = digitos[::-1]

def permutar(lista, inicio):
    if inicio == len(lista):

        num = 0
        for d in lista:
            num = num * 10 + d

        print("\nPermutação:", num)

        div = 10

        while div <= num:
            esquerda = num // div
            direita = num % div

            print("Teste:", esquerda, "+", direita, "=", esquerda + direita)

            if esquerda + direita == n:
                print(">>> SOLUÇÃO ENCONTRADA")

            div *= 10

    else:
        for i in range(inicio, len(lista)):
            lista[inicio], lista[i] = lista[i], lista[inicio]
            permutar(lista, inicio + 1)
            lista[inicio], lista[i] = lista[i], lista[inicio]

permutar(digitos, 0)
```
