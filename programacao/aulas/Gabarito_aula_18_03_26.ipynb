```python
def busca_linear(lista, alvo):
    for i, valor in enumerate(lista):
        if valor == alvo:
            return i
    return -1
```

```python
lista = [34, 12, 5, 78, 23]

pos = busca_linear(lista, 23)

if pos != -1:
    print(f"Encontrado na posição {pos}")
else:
    print("Valor não encontrado")
```

```python
def busca_binaria(lista, alvo):
    inicio = 0
    fim = len(lista) - 1

    while inicio <= fim:
        meio = (inicio + fim) // 2

        if lista[meio] == alvo:
            return meio
        elif lista[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio - 1

    return -1
```

```python
lista = [34, 12, 5, 78, 23]

lista.sort()

pos = busca_binaria(lista, 23)

if pos != -1:
    print(f"Encontrado na posição {pos}")
else:
    print("Valor não encontrado")
```

```python
# ATIVIDADE DO DIA
# Criar um sistema em Python com menu que permita:
# 1. Cadastrar números
# 2. Listar números
# 3. Ordenar (Bubble, Selection, Insertion)
# 4. Buscar (Linear e Binária)
# 5. Remover números
# 6. Sair

def bubble_sort(lista):
    n = len(lista)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]


def selection_sort(lista):
    n = len(lista)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if lista[j] < lista[min_idx]:
                min_idx = j
        lista[i], lista[min_idx] = lista[min_idx], lista[i]


def insertion_sort(lista):
    for i in range(1, len(lista)):
        chave = lista[i]
        j = i - 1
        while j >= 0 and lista[j] > chave:
            lista[j + 1] = lista[j]
            j -= 1
        lista[j + 1] = chave


def busca_linear(lista, valor):
    for i in range(len(lista)):
        if lista[i] == valor:
            return i
    return -1


def busca_binaria(lista, valor):
    inicio = 0
    fim = len(lista) - 1

    while inicio <= fim:
        meio = (inicio + fim) // 2

        if lista[meio] == valor:
            return meio
        elif lista[meio] < valor:
            inicio = meio + 1
        else:
            fim = meio - 1

    return -1


# ================= MENU =================

lista = []

while True:
    print("\n===== MENU =====")
    print("1 - Cadastrar número")
    print("2 - Listar números")
    print("3 - Ordenar lista")
    print("4 - Buscar número")
    print("5 - Remover número")
    print("0 - Sair")

    opcao = input("Escolha: ")

    # CADASTRAR
    if opcao == "1":
        num = int(input("Digite um número: "))
        lista.append(num)
        print("Número cadastrado!")

    # LISTAR
    elif opcao == "2":
        print("Lista:", lista)

    # ORDENAR
    elif opcao == "3":
        print("\n1 - Bubble Sort")
        print("2 - Selection Sort")
        print("3 - Insertion Sort")
        tipo = input("Escolha o tipo: ")

        if tipo == "1":
            bubble_sort(lista)
        elif tipo == "2":
            selection_sort(lista)
        elif tipo == "3":
            insertion_sort(lista)

        print("Lista ordenada:", lista)

    # BUSCAR
    elif opcao == "4":
        valor = int(input("Digite o número para buscar: "))

        print("\n1 - Busca Linear")
        print("2 - Busca Binária")
        tipo = input("Escolha: ")

        if tipo == "1":
            pos = busca_linear(lista, valor)

        elif tipo == "2":
            # garante ordenação
            lista.sort()
            pos = busca_binaria(lista, valor)
            print("Lista ordenada automaticamente:", lista)

        if pos != -1:
            print(f"Encontrado na posição {pos}")
        else:
            print("Não encontrado")

    # REMOVER
    elif opcao == "5":
        valor = int(input("Digite o número para remover: "))

        if valor in lista:
            lista.remove(valor)
            print("Removido com sucesso!")
        else:
            print("Número não está na lista.")

    # SAIR
    elif opcao == "0":
        print("Encerrando...")
        break

    else:
        print("Opção inválida!")
```
